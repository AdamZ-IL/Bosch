```python
import dask.dataframe as dd
#import graphviz as gv

df = dd.read_csv('../Data/train_numeric.csv')

thing = df.rtruediv(df).fillna(0)

colnames = list(df.columns.values[1:969])
type(colnames)
```

```python
print(df.visualize)
```

```python
df.get_partition(1)
```

```python
grouped = thing.groupby(colnames, method='disk', as_index=False).count().compute()




```

```python


grouped_small = thing.groupby(colnames[0:10], method='disk', as_Index=False).count().compute()
```

```python
grouped.reset_index().to_csv('../Data/train_numeric_grouped.csv', index=False)
```
